### Step 1. Import the necessary libraries

```python
import pandas as pd 
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
users.head()
users.isna()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")["age"].mean().reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male_count = users[users["gender"] == "M"].groupby("occupation").size()
total_count = users.groupby("occupation").size()
male_ratio = male_count / total_count 
male_ratio_df = male_ratio.reset_index(name = "Male Ratio").sort_values(by ="Male Ratio" ,ascending=False)
print(male_ratio_df)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")['age'].agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender'])["age"].agg(['mean'])
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_counts = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
gender_counts['Total'] = gender_counts.sum(axis=1)

gender_counts['Female percentage'] = (gender_counts.get('F',0)/gender_counts['Total']) * 100
gender_counts['Male percentage'] = (gender_counts.get('M',0)/ gender_counts['Total']) * 100

percentagess_df = gender_counts[['Female percentage','Male percentage']].reset_index()
#print(percentagess_df)
```
